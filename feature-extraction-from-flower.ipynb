```python
import scipy
import utils
import cv2, requests
import numpy as np
import numpy
import tensorflow.python.platform
import tensorflow as tf
from scipy import misc
import os
```

```python
def load_image_array(image_file):
	img = misc.imread(image_file)
	# GRAYSCALE
	if len(img.shape) == 2:
		img_new = np.ndarray( (img.shape[0], img.shape[1], 3), dtype = 'float32')
		img_new[:,:,0] = img
		img_new[:,:,1] = img
		img_new[:,:,2] = img
		img = img_new

	img_resized = misc.imresize(img, (224, 224))
	return (img_resized/255.0).astype('float32')
```

```python
vgg_file = open('/home/shivam/ML/vgg16-20160129.tfmodel')
vgg16raw = vgg_file.read()
vgg_file.close()

graph_def = tf.GraphDef()
graph_def.ParseFromString(vgg16raw)
images = tf.placeholder("float32", [None, 224, 224, 3])
tf.import_graph_def(graph_def, input_map={ "images": images })
graph = tf.get_default_graph()
sess = tf.Session()

arr = []

for root, dirs, files in os.walk("/home/shivam/ML/jpg/"):
    for file in files:
        print file
        image_array = load_image_array("jpg/" + file)
        image_feed = np.ndarray((1,224,224,3))
        image_feed[0:,:,:] = image_array
        feed_dict  = { images : image_feed }
        fc7_tensor = graph.get_tensor_by_name("import/fc7/BiasAdd:0")
        fc7_features = sess.run(fc7_tensor, feed_dict = feed_dict)
        arr.append(fc7_features)
        #print fc7_features
sess.close()
#print fc7_features
```
