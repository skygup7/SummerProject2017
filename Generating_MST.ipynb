```python
import numpy as np
import pandas as pd
from scipy.spatial import distance
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import minimum_spanning_tree
```

```python
df= pd.read_csv('dataframe_pca.csv', index_col=0)
df
```

```python
old= df.values
old
```

```python
arr = np.zeros((8189,8189), dtype=np.float)

for i in range(0,8189):
    for j in range(i,8189):
        arr[i][j] = distance.euclidean(old[i],old[j])
```

```python
df2 = pd.DataFrame(data=arr)
```

```python
df2.to_csv('distance_uppertr.csv')
```

```python
X = csr_matrix(arr)
```

```python
Tcsr = minimum_spanning_tree(X)
```

```python
print(Tcsr)
```

```python
df3 = pd.DataFrame(data=Tcsr.toarray().astype(float))
```

```python
df3.to_csv('minimum_spanning_tree.csv')
df3
```
