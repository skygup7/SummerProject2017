```python
df = pd.read_csv('../distance_uppertr.csv', index_col=0)
X = csr_matrix(df.values)
Tcsr = minimum_spanning_tree(X)
```

```python
from scipy.sparse import csgraph
```

```python
X = csgraph.laplacian(Tcsr)
```
